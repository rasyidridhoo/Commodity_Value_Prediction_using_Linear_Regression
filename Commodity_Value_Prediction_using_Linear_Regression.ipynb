# Data Preprocessing

```python
X = df.drop("harga tahun 2022", axis=1)
y = df["harga tahun 2022"]  
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
import math
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=0)
```

# Modelling

```python
from sklearn.linear_model import LinearRegression 
Reg = LinearRegression()
```

# Predict with Linear Regression

```python
Reg.fit(X_train,y_train).predict(X_test)
```

```python
preds_2 = Reg.predict(X_test)
preds_round = []
for x in preds_2:
  preds_round.append(math.ceil(x))

preds_round
```

# Calculation Accuracy training data and test data

```python
accuracy_train = Reg.score(X_train, y_train)
print('Accuracy for Training: ', accuracy_train)
```

```python
accuracy_test = Reg.score(X_test, y_test)
print('Accuracy for Test: ',accuracy_test)
```
